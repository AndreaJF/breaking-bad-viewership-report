# Viewership Over Time

```python
data = {
    'Season': [1, 2, 3, 4, 5],
    'Viewers (millions)': [1.23, 1.3, 1.52, 1.9, 4.32]  
}
```

Data source: https://quirkyfigures.com/en/average-viewers-per-episode-by-season-in-breaking-bad-tv-series

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Season'], df['Viewers (millions)'], marker='o')
plt.title('Viewership of Breaking Bad Over Time')
plt.xlabel('Season')
plt.ylabel('Viewers (millions)')
plt.grid(True)
plt.show()
```

```python
data1 = {
    'Season': [1, 2, 3, 4, 5],  
    'Viewers (millions)': [1.23, 1.3, 1.52, 1.9, 4.32]  
}
```

```python
df1 = pd.DataFrame(data1)
```

```python
df1['Viewership Change'] = df1['Viewers (millions)'].diff()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df1['Season'], df1['Viewership Change'], marker='o', color='green')
plt.title('Season-to-Season Changes in Viewership for Breaking Bad')
plt.xlabel('Season')
plt.ylabel('Viewership Change (millions)')
plt.grid(True)
plt.show()
```

The viewership for "Breaking Bad" experienced significant growth over its five seasons. In Season 1, the viewership started at 1.23 million and increased to 1.3 million in Season 2, reflecting a modest increase of 0.07 million viewers. However, the viewership saw a more substantial increase in Season 3, reaching 1.52 million viewers, indicating a notable increase of 0.22 million viewers compared to Season 2.

This upward trend continued in Season 4, with viewership climbing to 1.9 million viewers, representing a significant increase of 0.38 million viewers compared to Season 3. The most significant surge in viewership occurred in the final season, Season 5, where viewership skyrocketed to 4.32 million viewers, marking a remarkable increase of 2.42 million viewers compared to Season 4.
